# Rotation Gates: $R_x$, $R_y$, $R_z$ - Introduction and Mathematics

## Introduction
The rotation gates ($R_x$, $R_y$, $R_z$) are a family of single-qubit gates that apply a rotation to a qubit around the $x$-, $y$-, or $z$-axis of the Bloch sphere. These gates are important in quantum computing because they allow for precise control over the quantum state of a qubit. They are especially useful in quantum algorithms that require specific rotations to manipulate qubit states.

### General Rotation Gate
A rotation gate on a single qubit can be represented as:
$$
R_{\theta} = e^{-i \frac{\theta}{2} \hat{n} \cdot \sigma},
$$
where $ \theta $ is the angle of rotation, $\hat{n}$ is the unit vector indicating the axis of rotation (e.g., $x$, $y$, or $z$), and $\sigma$ is the Pauli vector of operators $(\sigma_x, \sigma_y, \sigma_z)$ depending on the axis of rotation.

---

### $R_x$ Gate (Rotation about the X-axis)

The $R_x(\theta)$ gate applies a rotation around the $x$-axis of the Bloch sphere. Its matrix representation is:

$$
R_x(\theta) = e^{-i \frac{\theta}{2} X} = \cos\left(\frac{\theta}{2}\right) I - i \sin\left(\frac{\theta}{2}\right) X,
$$

where $X$ is the Pauli-X matrix:
$$
X = \begin{bmatrix} 
0 & 1 \\
1 & 0
\end{bmatrix}.
$$

The action of the $R_x$ gate on the basis states is:

- $$ R_x(\theta) \ket{0} = \cos\left(\frac{\theta}{2}\right) \ket{0} - i \sin\left(\frac{\theta}{2}\right) \ket{1} $$
- $$ R_x(\theta) \ket{1} = \cos\left(\frac{\theta}{2}\right) \ket{1} - i \sin\left(\frac{\theta}{2}\right) \ket{0} $$

---

### $R_y$ Gate (Rotation about the Y-axis)

The $R_y(\theta)$ gate applies a rotation around the $y$-axis of the Bloch sphere. Its matrix representation is:

$$
R_y(\theta) = e^{-i \frac{\theta}{2} Y} = \cos\left(\frac{\theta}{2}\right) I - i \sin\left(\frac{\theta}{2}\right) Y,
$$

where $Y$ is the Pauli-Y matrix:
$$
Y = \begin{bmatrix} 
0 & -i \\
i & 0
\end{bmatrix}.
$$

The action of the $R_y$ gate on the basis states is:

- $$ R_y(\theta) \ket{0} = \cos\left(\frac{\theta}{2}\right) \ket{0} - i \sin\left(\frac{\theta}{2}\right) \ket{1} $$
- $$ R_y(\theta) \ket{1} = \cos\left(\frac{\theta}{2}\right) \ket{1} + i \sin\left(\frac{\theta}{2}\right) \ket{0} $$

---

### $R_z$ Gate (Rotation about the Z-axis)

The $R_z(\theta)$ gate applies a rotation around the $z$-axis of the Bloch sphere. Its matrix representation is:

$$
R_z(\theta) = e^{-i \frac{\theta}{2} Z} = \cos\left(\frac{\theta}{2}\right) I - i \sin\left(\frac{\theta}{2}\right) Z,
$$

where $Z$ is the Pauli-Z matrix:
$$
Z = \begin{bmatrix} 
1 & 0 \\
0 & -1
\end{bmatrix}.
$$

The action of the $R_z$ gate on the basis states is:

- $$ R_z(\theta) \ket{0} = \cos\left(\frac{\theta}{2}\right) \ket{0} - i \sin\left(\frac{\theta}{2}\right) \ket{0} $$
- $$ R_z(\theta) \ket{1} = \cos\left(\frac{\theta}{2}\right) \ket{1} + i \sin\left(\frac{\theta}{2}\right) \ket{1} $$

---

### Summary of Rotation Gates

- The $R_x$ gate rotates a qubit around the $x$-axis.
- The $R_y$ gate rotates a qubit around the $y$-axis.
- The $R_z$ gate rotates a qubit around the $z$-axis.

Each of these gates can be used to manipulate a qubitâ€™s state on the Bloch sphere by specifying the angle of rotation ($\theta$). They are fundamental for precise control in quantum algorithms and are typically used in combination with other gates to achieve complex state transformations.

## $R_x$

```python
qc = QuantumCircuit(1, 1)

qc.x(0)

qc.rx(pi/2,0)

qc.measure(0, 0)

simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit)
result = job.result()


counts = result.get_counts()


plt.bar(counts.keys(), counts.values(), color='blue')
plt.title("Measurement Histogram")
plt.show()

print("Quantum Circuit")
qc.draw('mpl')
```

# Applications and Exercises for Rotation Gates ($R_x$, $R_y$, $R_z$)

---

## **Applications**

1. **State Preparation**:
   - Rotation gates are essential for preparing qubits in specific superpositions. For instance:
     - $R_y$ can prepare arbitrary qubit states $\cos(\theta/2)\ket{0} + \sin(\theta/2)\ket{1}$.
     - $R_x$ and $R_z$ are used to adjust the phases or amplitudes of states.

2. **Quantum Machine Learning**:
   - Rotation gates are frequently used in variational quantum algorithms like the Variational Quantum Eigensolver (VQE) and Quantum Neural Networks (QNN). These gates help adjust the parameters of quantum circuits to optimize the cost function.

3. **Quantum Cryptography**:
   - In protocols like Quantum Key Distribution (QKD), rotation gates play a key role in encoding and decoding quantum states, ensuring secure communication.

4. **Quantum Simulation**:
   - Rotation gates are fundamental for simulating quantum systems, particularly when applying the operators representing time evolution or Hamiltonians.

## **Exercises**

### **Exercise 1: Rotation of a Qubit Using $R_x$**
- **Task**: Apply the $R_x$ gate to a qubit in the state $\ket{0}$. What is the resulting state if $\theta = \pi/2$?

- **Solution**:
   - Apply the $R_x$ gate:
     $$
     R_x(\theta) \ket{0} = \cos(\theta/2) \ket{0} + i \sin(\theta/2) \ket{1}
     $$
   - Substituting $\theta = \pi/2$:
     $$
     R_x(\pi/2) \ket{0} = \cos(\pi/4) \ket{0} + i \sin(\pi/4) \ket{1} = \frac{1}{\sqrt{2}} (\ket{0} + i \ket{1})
     $$

### **Exercise 2: Rotation of a Qubit Using $R_y$**
- **Task**: Apply the $R_y$ gate to a qubit in the state $\ket{0}$. What is the resulting state if $\theta = \pi$?

- **Solution**:
   - Apply the $R_y$ gate:
     $$
     R_y(\theta) \ket{0} = \cos(\theta/2) \ket{0} + \sin(\theta/2) \ket{1}
     $$
   - Substituting $\theta = \pi$:
     $$
     R_y(\pi) \ket{0} = \cos(\pi/2) \ket{0} + \sin(\pi/2) \ket{1} = \ket{1}
     $$

### **Exercise 3: Rotation of a Qubit Using $R_z$**
- **Task**: Apply the $R_z$ gate to a qubit in the state $\ket{0}$. What is the resulting state if $\theta = \pi/2$?

- **Solution**:
   - Apply the $R_z$ gate:
     $$
     R_z(\theta) \ket{0} = e^{-i\theta/2} \ket{0}
     $$
   - Substituting $\theta = \pi/2$:
     $$
     R_z(\pi/2) \ket{0} = e^{-i\pi/4} \ket{0} = \frac{1}{\sqrt{2}} (1 - i) \ket{0}
     $$

### **Exercise 4: Composite Rotation**
- **Task**: Apply a sequence of rotations $R_y(\theta)$, $R_x(\phi)$, and $R_z(\gamma)$ to a qubit in the state $\ket{0}$. What is the final state?

- **Solution**:
   - First, apply $R_y(\theta)$ to $\ket{0}$:
     $$
     R_y(\theta) \ket{0} = \cos(\theta/2) \ket{0} + \sin(\theta/2) \ket{1}
     $$
   - Then, apply $R_x(\phi)$:
     $$
     R_x(\phi) \left( \cos(\theta/2) \ket{0} + \sin(\theta/2) \ket{1} \right)
     $$
     This will result in a superposition, and finally, apply $R_z(\gamma)$ to the resulting state:
     $$
     R_z(\gamma) \left( \cos(\theta/2) \ket{0} + \sin(\theta/2) \ket{1} \right)
     $$

   The final state will be a linear combination of $\ket{0}$ and $\ket{1}$, with the corresponding coefficients based on $\theta$, $\phi$, and $\gamma$.
