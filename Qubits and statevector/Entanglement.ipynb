## Entanglement and Multi-Qubit Systems  

Entanglement is a unique quantum phenomenon that occurs when the state of one qubit becomes inseparably linked to the state of another, even across large distances. To understand entanglement and multi-qubit systems, we explore key concepts such as tensor products, Bell states, and quantum correlations.  

---

### Tensor Product and Combined State Representation  

When dealing with multiple qubits, the combined state of the system is represented using the **tensor product** of individual qubit states.  

#### Single-Qubit States  
Suppose we have two qubits with the states:  
- Qubit 1: $|\psi_1\rangle = \alpha|0\rangle + \beta|1\rangle$  
- Qubit 2: $|\psi_2\rangle = \gamma|0\rangle + \delta|1\rangle$  

The combined state of the two-qubit system is represented as:  
$$ 
|\Psi\rangle = |\psi_1\rangle \otimes |\psi_2\rangle = (\alpha|0\rangle + \beta|1\rangle) \otimes (\gamma|0\rangle + \delta|1\rangle).
$$  

Using the distributive property of the tensor product, this expands to:  
$$ 
|\Psi\rangle = \alpha\gamma|00\rangle + \alpha\delta|01\rangle + \beta\gamma|10\rangle + \beta\delta|11\rangle,
$$  
where $|00\rangle$, $|01\rangle$, $|10\rangle$, and $|11\rangle$ represent the basis states of the two-qubit system.  

#### Corresponding Operation  
For two vectors:  
- $|\psi_1\rangle = \begin{bmatrix} a \\ b \end{bmatrix}, \quad |\psi_2\rangle = \begin{bmatrix} c \\ d \end{bmatrix}$,  
the tensor product is computed as:  
$$
|\Psi\rangle = |\psi_1\rangle \otimes |\psi_2\rangle = \begin{bmatrix} a \\ b \end{bmatrix} \otimes \begin{bmatrix} c \\ d \end{bmatrix} = \begin{bmatrix} ac \\ ad \\ bc \\ bd \end{bmatrix}.
$$

### Bell States and Their Significance  

The **Bell states** are specific examples of maximally entangled two-qubit states. They form an orthonormal basis for the two-qubit system and are given as:  

1. $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$  
2. $|\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle)$  
3. $|\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle)$  
4. $|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle)$  

#### Properties of Bell States  
- They are **entangled**, meaning the measurement of one qubit affects the other.  
- Measurement outcomes are perfectly correlated or anti-correlated, depending on the Bell state.  

#### Corresponding Operation  
- To create a Bell state $|\Phi^+\rangle$ from two qubits initially in $|00\rangle$:  
  1. Apply a **Hadamard gate** to the first qubit:  
     $$H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle).$$  
  2. Apply a **CNOT gate** with the first qubit as the control and the second as the target:  
     $$CNOT\left(\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \otimes |0\rangle\right) = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).$$  

---

### Entanglement and Quantum Correlations  

#### Definition of Entanglement  
A multi-qubit state is **entangled** if it cannot be written as the tensor product of individual qubit states. For example:  
- The state $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$ is entangled because it cannot be separated into two independent single-qubit states.  
- In contrast, $|\psi\rangle = (\alpha|0\rangle + \beta|1\rangle) \otimes (\gamma|0\rangle + \delta|1\rangle)$ is **separable**, as it is a tensor product of two states.  

#### Quantum Correlations  
Entangled states exhibit quantum correlations, meaning the outcome of a measurement on one qubit is related to the outcome of the other, even if they are spatially separated. These correlations violate classical expectations, as described by **Bell's theorem**.  

#### Corresponding Operation  
- Measure each qubit in the Bell state $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$:  
  - If the first qubit is measured as $|0\rangle$, the second qubit will also be $|0\rangle$.  
  - If the first qubit is measured as $|1\rangle$, the second qubit will also be $|1\rangle$.  

---

### Key Takeaways  

1. The **tensor product** combines the states of multiple qubits, forming a higher-dimensional state space.  
2. **Bell states** are examples of maximally entangled states and are critical for quantum information tasks like teleportation and superdense coding.  
3. **Entanglement** enables quantum correlations, forming the basis for phenomena that defy classical intuition, such as non-locality.
