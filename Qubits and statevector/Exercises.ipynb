## Exercise 1: Tensor Product Calculation
Given the states:  
- $|\psi_1\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$
- $|\psi_2\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$

1. Compute the tensor product $|\psi_1\rangle \otimes |\psi_2\rangle$.  
2. What is the basis state representation of the resulting vector?

### Solution
1. **Tensor Product**  
   The tensor product is computed as:  
   $$
   |\psi_1\rangle \otimes |\psi_2\rangle = 
   \begin{bmatrix} 1 \\ 0 \end{bmatrix} \otimes \begin{bmatrix} 0 \\ 1 \end{bmatrix}
   = \begin{bmatrix} 1 \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} \\ 0 \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} \end{bmatrix}
   = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}.
   $$

2. **Basis State Representation**  
   The resulting vector corresponds to the basis state $|01\rangle$ in the combined two-qubit system.

## Exercise 2: Bell State Verification
The Bell state $|\Phi^+\rangle$ is defined as:  
$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).
$$

1. Write $|\Phi^+\rangle$ in vector form.  
2. Verify that $|\Phi^+\rangle$ is normalized.

### Solution
1. **Vector Form of $|\Phi^+\rangle$**  
   The state $|\Phi^+\rangle$ can be written as:  
   $$
   |\Phi^+\rangle = \frac{1}{\sqrt{2}} 
   \begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}
   + \frac{1}{\sqrt{2}} 
   \begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}
   = \frac{1}{\sqrt{2}}
   \begin{bmatrix} 1 \\ 0 \\ 0 \\ 1 \end{bmatrix}.
   $$

2. **Normalization**  
   To check normalization, calculate the norm:  
   $$
   \langle\Phi^+|\Phi^+\rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 0 & 0 & 1 \end{bmatrix} 
   \cdot \frac{1}{\sqrt{2}} 
   \begin{bmatrix} 1 \\ 0 \\ 0 \\ 1 \end{bmatrix}
   = \frac{1}{2} (1^2 + 0^2 + 0^2 + 1^2) = \frac{1}{2}(2) = 1.
   $$
   Hence, $|\Phi^+\rangle$ is normalized.

## Exercise 3: Entanglement Test
Given the state:  
$$
|\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle),
$$
determine if $|\psi\rangle$ is entangled.

### Solution
1. **Check for Factorization**  
   If $|\psi\rangle$ can be written as $|\psi_1\rangle \otimes |\psi_2\rangle$, it is separable (not entangled).  
   Assume $|\psi\rangle = (\alpha|0\rangle + \beta|1\rangle) \otimes (\gamma|0\rangle + \delta|1\rangle)$:
   $$
   |\psi\rangle = \alpha\gamma|00\rangle + \alpha\delta|01\rangle + \beta\gamma|10\rangle + \beta\delta|11\rangle.
   $$
   Comparing with $|\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle)$, note that the $|01\rangle$ and $|11\rangle$ terms are missing.  
   Thus, $|\psi\rangle$ cannot be factorized, and it is entangled.

## Exercise 4: Check if a State is Entangled
Given the state:
$$
|\psi\rangle = \frac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle),
$$
1. Check if this state is separable (not entangled).
2. If separable, express it as a tensor product of two single-qubit states. If entangled, explain why.

### Solution
1. **Factorization Check**  
   Assume $|\psi\rangle = (\alpha|0\rangle + \beta|1\rangle) \otimes (\gamma|0\rangle + \delta|1\rangle)$:
   $$
   |\psi\rangle = \alpha\gamma|00\rangle + \alpha\delta|01\rangle + \beta\gamma|10\rangle + \beta\delta|11\rangle.
   $$
   Compare this with the given state:
   $$
   |\psi\rangle = \frac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle).
   $$

   As we can match the coefficients of all terms, the state is **separable** and not entangled. In fact, the state is just an equal superposition of all four computational basis states, which is a product state.

## Exercise 5: Constructing a Bell State from Single-Qubit States
Given two qubits in the states:
- $|\alpha\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$
- $|\beta\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$

1. Construct the Bell state $|\Phi^+\rangle$ by taking the tensor product of $|\alpha\rangle$ and $|\beta\rangle$.
2. Verify if the resulting state is normalized.

### Solution
1. **Constructing the Bell State**  
   Compute the tensor product:
   $$
   |\Phi^+\rangle = |\alpha\rangle \otimes |\beta\rangle
   = \left(\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)\right) \otimes \left(\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)\right)
   = \frac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle).
   $$

2. **Normalization Check**  
   To check normalization, calculate the norm:
   $$
   \langle \Phi^+ | \Phi^+ \rangle = \frac{1}{4} (1^2 + 1^2 + 1^2 + 1^2) = 1.
   $$

   Therefore, the state $|\Phi^+\rangle$ is normalized.

## Exercise 6: EPR Paradox
In the EPR paradox, two particles are entangled in the state:
$$
|\psi_{\text{EPR}}\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).
$$

1. If we measure the first qubit and obtain $|0\rangle$, what is the state of the second qubit?
2. If we measure the second qubit and obtain $|1\rangle$, what is the state of the first qubit?

### Solution
1. **Measurement of the First Qubit**  
   If the first qubit is measured and found in the state $|0\rangle$, the overall state collapses to:
   $$
   |\psi_{\text{EPR}}\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \rightarrow |0\rangle \otimes |0\rangle.
   $$

   Hence, the second qubit is in the state $|0\rangle$.

2. **Measurement of the Second Qubit**  
   If the second qubit is measured and found in the state $|1\rangle$, the overall state collapses to:
   $$
   |\psi_{\text{EPR}}\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \rightarrow |1\rangle \otimes |1\rangle.
   $$

   Therefore, the first qubit is in the state $|1\rangle$.

## Exercise 7: Bell-State Measurement
Given the two-qubit state:
$$
|\psi_{\text{entangled}}\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle),
$$
perform a Bell-state measurement on this state.

1. Calculate the measurement outcomes.
2. What is the probability of obtaining the outcome $|\Phi^+\rangle$?

### Solution
1. **Bell-State Measurement**  
   A Bell-state measurement involves projecting the state onto one of the four Bell states. The four Bell states are:
   $$
   |\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle),
   |\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle),
   |\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle),
   |\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle).
   $$

   Since $|\psi_{\text{entangled}}\rangle$ is already in the state $|\Phi^+\rangle$, the outcome of the Bell-state measurement will always be $|\Phi^+\rangle$.

2. **Probability of Outcome**  
   The probability of measuring $|\Phi^+\rangle$ is:
   $$
   P(|\Phi^+\rangle) = |\langle \Phi^+|\psi_{\text{entangled}}\rangle|^2 = 1.
   $$

   Therefore, the probability of obtaining $|\Phi^+\rangle$ is 100%.

## Exercise 8: Entanglement Detection
Consider the two-qubit state:
$$
|\psi_{\text{state}}\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle).
$$

1. Is this state entangled?
2. What is the tensor product decomposition of this state if it is separable?

### Solution
1. **Check for Entanglement**  
   We can try to factorize the state into a product of two single-qubit states.  
   Assume $|\psi_{\text{state}}\rangle = (\alpha|0\rangle + \beta|1\rangle) \otimes (\gamma|0\rangle + \delta|1\rangle)$:
   $$
   |\psi_{\text{state}}\rangle = \alpha\gamma|00\rangle + \alpha\delta|01\rangle + \beta\gamma|10\rangle + \beta\delta|11\rangle.
   $$
   Comparing with $|\psi_{\text{state}}\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle)$, we see that the state cannot be factorized into a product of two single-qubit states. Therefore, $|\psi_{\text{state}}\rangle$ is **entangled**.

2. **Separable Decomposition**  
   Since the state is entangled, there is no separable tensor product decomposition.
