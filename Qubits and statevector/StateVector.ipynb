# State Vector Representation

## Basis States 0 and 1
In quantum mechanics, the states 0 and 1 are called the **computational basis states**. They serve as the building blocks for describing the state of a qubit.

Mathematically, these basis states can be represented as column vectors:
$$
|0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad
|1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}.
$$

Any qubit state can be described as a **linear combination** (or superposition) of these basis states.

---

## Superposition Principle
A qubit can exist in a superposition of 0 and 1, which means it can be in a state that is a weighted combination of both basis states. The general form of a qubit state is:
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$$
where:
- α and β are complex numbers.
- α^2 and β^2 represent the probabilities of measuring the qubit in 0 or 1, respectively.
- The normalization condition ensures that the probabilities sum to 1:
  $$
  |\alpha|^2 + |\beta|^2 = 1.
  $$

For example:
- If $$ |\psi\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle $$, the probabilities of 0 and 1 are equal (\( 50\% \) each).
- If $$ |\psi\rangle = |0\rangle , \text{the state is fully in} |0\rangle $$ with \( 100\% \) probability.

---

## Representation Using Dirac Notation
In **Dirac notation**, quantum states are written as Ψ (called a **ket**) and are elements of a complex vector space. The dual of Ψ is Ψ called a **bra**), which represents the conjugate transpose of the state vector.

For a state $$ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle $$, the dual is:
$$
\langle\psi| = \alpha^* \langle 0| + \beta^* \langle 1|,
$$
where α^* and β^* are the complex conjugates of α and β, respectively.

The inner product between two states Ψ and Ψ is:
$$
\langle\phi|\psi\rangle = \sum_i \phi_i^* \psi_i.
$$

## Bloch Sphere Visualization
The **Bloch sphere** is a geometrical representation of a qubit's state. It provides an intuitive way to visualize the state of a single qubit.

### Parametrization of the State
A general qubit state can also be written as:
$$
|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\right)|1\rangle,
$$
where: 
 $$ \theta \text{ (polar angle) and } \phi \text{ (azimuthal angle) define the position of the state on the Bloch sphere.}$$ 
 $$ \cos^2\left(\frac{\theta}{2}\right) \text{is the probability of measuring} |0\rangle$$
 $$ \sin^2\left(\frac{\theta}{2}\right) \text{is the probability of measuring} |1\rangle$$

### Bloch Sphere Features
- The north pole θ = 0 represents 0.
- The south pole θ = π represents 1.
- Points on the equator θ = π/2 are equal superpositions of 0 and 1, with different relative phases determined by Φ.

The Bloch sphere is described by the equation:
$$
x = \sin\theta \cos\phi, \quad y = \sin\theta \sin\phi, \quad z = \cos\theta,
$$
where \( x, y, z \) are the Cartesian coordinates.

---

### Visualization
On the Bloch sphere:
- The \( z \)-axis represents the probabilities of 0 and 1.
- The \( x \)- and \( y \)-axes represent the phase relationships between 0 and 1.

The Bloch sphere is a powerful tool for understanding single-qubit operations and visualizing the effects of quantum gates.
